```python
#Food Security Assessment

meals_per_day = [3, 2, 1, 3, 2, 2, 1]
for meals in meals_per_day:

    if meals >= 3:
        print(meals, "meals - Food secure")

    elif meals == 2:
        print(meals, "meals - Moderately food insecure")

    else:

        print(meals, "meals - Severely food insecure")
```

```python
#Refugee Registration Analysis
# Ages of newly registered refugees
refugee_ages = [8, 45, 62, 3, 17, 28, 71, 12, 34, 5, 55, 2, 19,]

for age in refugee_ages:
    if 0 <= age <= 4:
        print(age, "years - Early Childhood Program")

    elif 5 <= age <= 17:
        print(age, "years - Education Program")

    elif 18 <= age <= 59:
       print(age, "years - Livelihood Program")

    elif age >= 60:
        print(age, "years - Elderly Care Program")
```

```python
water_access = [12, 28, 18, 10, 35, 22, 15, 8, 30, 19, 14, 26]

# Variable to count critical shortage cases
critical_shortage = 0

# Loop through each household's water access
for water in water_access:
    if water < 15:
        classification = "Critical shortage - needs immediate intervention"
        critical_shortage += 1  # Increment counter
    elif 15 <= water <= 25:
        classification = "Below standard - needs improvement"
    else:
        classification = "Adequate access"

    # Print each household's water access and classification
    print(f"{water} liters per person: {classification}")

# Print total number of households with critical shortage
print(f"\nTotal households with critical shortage: {critical_shortage}")
```
